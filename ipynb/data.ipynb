```python
from collections import defaultdict
```

```python
with open('/home/kolegor/Downloads/trainingData.csv', 'r') as inf:
    languages = defaultdict(int)
    for line in inf:
        break
    for line in inf:
        name, language = line.split(',')
        languages[language.strip()] += 1
```

```python
for name, count in sorted(languages.iteritems()):
    print name, count
```
